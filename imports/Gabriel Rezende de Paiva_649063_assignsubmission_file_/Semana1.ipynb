# Exercício 4

Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas na semana. Calcule e mostre o total do seu salário ao final de 4 semanas. 

```python
valor = float(input("Digite quanto você ganha por hora"))
horas = float(input("Digite quantas horas você trabalha por semana"))

salario = (valor * horas) * 4

textosalario = f"R${salario:_.2f}"
textosalario = textosalario.replace(".", ",").replace("_",".")

print(f"O total do seu salário ao final de 4 semanas é de {textosalario}")
```

# Exercício 8

Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 12 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R$ 80,00 ou em galões de 3,6 litros, que custam R$ 25,00.

Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:

    - comprar apenas latas de 18 litros;
    - comprar apenas galões de 3,6 litros;
    - misturar latas e galões, de forma que o desperdício de tinta seja menor. 
    
Acrescente 10% de folga nos três casos (isto é, você deve adicionar 10% sobre o total de litros de tinta necessário) e sempre arredonde os valores para cima, isto é, considere latas cheias. 



```python
import math

area = float(input('Digite a área em metros quadrados a ser pintada'))
folga = area * 1.1
litros_por_metro = 12
litros_usados = folga / litros_por_metro
litros_lata = 18
quantidade_latas = math.ceil(litros_usados / litros_lata)
valor_so_latas = quantidade_latas * 80
print(f'Comprando apenas latas de 18 litros você usará: {quantidade_latas} latas, totalizando {valor_so_latas} reais')

litros_galao = 3.6
quantidade_galoes = math.ceil(litros_usados / litros_galao)
valor_so_galoes = quantidade_galoes * 25
print(f'Comprando apenas galões de 3,6 litros você usará: {quantidade_galoes} galões, totalizando {valor_so_galoes} reais')

quantidade_latas = math.floor(litros_usados / litros_lata)
valor_de_latas = quantidade_latas * 80
litros_faltantes = litros_usados % litros_lata
quantidade_galoes=math.ceil(litros_faltantes / litros_galao)
valor_com_galoes = quantidade_galoes * 25

valor_total = valor_de_latas + valor_com_galoes

print(f'A melhor opção seria usar {quantidade_latas} latas de 18 litros e {quantidade_galoes} galões de 3.6 litros, totalizando {valor_total} reais')
```
